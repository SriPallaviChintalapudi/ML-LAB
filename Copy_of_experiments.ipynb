# 2. Write python program for consider a given vector, how to add 1 to each element indexed by a second vector (be careful with repeated indices)?

```python
import numpy as np

Z = np.ones(10)
print("Ones Array: ",Z)
length=len(Z)
print("Length: ",length)

I = np.random.randint(0,length,20)
print("Random Array: ",I)
I_length=len(I)
print("Length: ",I_length)

Z += np.bincount(I, minlength=len(Z))
print("Result: ",Z)
```

# 5. Write python program for given an arbitrary number of vectors, build the cartesian product (every combinations of every item).

```python
import numpy as np
x = np.array([1,2,3])
y = np.array([4,5])
res = np.transpose([np.tile(x, len(y)), np.repeat(y, len(x))])
print("Cartesian Product for Two Arrays:")
print(res)
```

```python
import numpy as np
def cartesian(arrays):
    arrays = [np.asarray(a) for a in arrays]
    print("Arrays: \n",arrays)
    
    shape = (len(x) for x in arrays)
    

    ix = np.indices(shape, dtype=int)
    ix = ix.reshape(len(arrays), -1).T
    for n, arr in enumerate(arrays):
        ix[:, n] = arrays[n][ix[:, n]]
        #print(ix)

    return ix

arr1=[1, 2, 3]
arr2=[4,5]
arr3=[6,7]
arr4=[8,9,0]
res=cartesian((arr1, arr2, arr3,arr4))
print("Cartesian Product for Multiple Arrays: \n", res)
```

# 6. Write python program to get value of x for given linear equations.

```python
from sympy import symbols, Eq, solve
x, y, z = symbols('x,y,z')
eq1 = Eq((x+y+z), 1)
print("Equation 1: ", eq1)
eq2 = Eq((x-y+2*z), 1)
print("Equation 2: ",eq2)
eq3 = Eq((2*x-y+2*z), 1)
print("Equation 3: ",eq3)
print("\n Values of 3 unknown variable are as follows:")
print(solve((eq1, eq2, eq3), (x, y, z)))
```

# 7. Write python program to find eigen analysis for given matrix.

```python
import numpy as np
m = np.array([[1, 2, 3],
              [2, 3, 4],
              [4, 5, 6]])
print("Printing the Original square array:\n",m)
w, v = np.linalg.eig(m)
print("Printing the Eigen values of the given square array:\n",w)
print("Printing Right eigenvectors of the given square array:\n",v)
```

```python
abalone = pd.get_dummies(abalone)
y = abalone["Rings"]
X = abalone.drop(columns="Rings")
```

```python
abalone['age'] = abalone['Rings']+1.5
abalone.drop('Rings', axis = 1, inplace = True)
abalone
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
from sklearn.neighbors import KNeighborsClassifier
# Initializing classifier and giving hyperparameter k=3
knn = KNeighborsClassifier(n_neighbors=3)
# training classifier
knn.fit(X_train, y_train)
# Evaluate the classifier
print(knn.score(X_test, y_test))
# Try changing hyperparameter
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X_train, y_train)
print(knn.score(X_test, y_test))
```

```python
from sklearn.linear_model import LogisticRegression
# Initializing classifier with one-v-rest approach. random_state is # to ensure same results in every execution.
logr = LogisticRegression(multi_class = 'ovr', random_state=3)
# training classifier
logr.fit(X_train, y_train)
# Evaluate the classifier
print(logr.score(X_test, y_test))
```

```python
from sklearn.tree import DecisionTreeClassifier
# Initializing classifier. random_state is to ensure same results in every execution. max_depth is to specify number of splits
dt = DecisionTreeClassifier(random_state=0, max_depth = 3)
# training classifier
dt.fit(X_train, y_train)
# Evaluate the classifier
print(dt.score(X_test, y_test))
```
